#### 3.2. An√°lisis de sentimientos
En este apartado realizaremos un peque√±o an√°lsis sobre nuestros tweets haciendo uso de la librer√≠a "textblob" (ya previamente importada), la cual constituye una herramienta perfecta para procesar texto de forma sencilla. Aunque tiene diversas funcionalidades, para el caso actual nos centraremos en la posiblidad que nos brinda esta librer√≠a para realizar un an√°lisis de sentimientos.
Empezaremos el proceso preparando los datos para el posterior an√°lisis con textblob, para ello crearemos una funci√≥n que haciendo uso del m√≥dulo de expresi√≥n regular "re" nos permita sustituir los caracteres especiales que puedan entorpecer por espacios en blanco.

```python
def clean_tweet(tweet):
    """Limpia los tweets removiendo caracteres especiales a trav√©s del m√≥dulo de expresiones regulares"""
    return ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", tweet).split())
```

Nuestro siguiente paso ser√° crear una funci√≥n que nos permita clasificar el sentimiento de los tweets en tres categor√≠as: "positivo", "neutral" y "negativo". TextBlob nos diferencia dos aspectos al darnos el "sentimiento" de una string cualquiera, la polaridad y la subjetividad; dado que la polaridad representa "el grado de positividad o negatividad" del texto analizado, decidimos pedir en nuestra funci√≥n la polaridad y tomarla como el valor del sentimiento de los tweets.
Por otra parte, hay que tener en cuenta que al no ser el idioma de los tweets recolectados el ingl√©s (ya que utilizamos keywords en castellano para hacer nuestra extracci√≥n de tweets), deberemos pasarle instrucciones concretas a la funci√≥n para que, en caso de encontrarse tweets en otro idioma, los traduzca al ingl√©s, ya que el funcionamiento b√°sico de la librer√≠a que estamos utilizando opera en ingl√©s.

```python
def analize_tweet_sentiment(tweet):
    """Analiza el sentimiento de los tweets y los clasifica en tres categor√≠as en funci√≥n de su polaridad"""
    try:
        analysis = TextBlob(clean_tweet(tweet))
        language = analysis.detect_language()
        if language == 'en':
            analysis_ready = analysis
        else:
            analysis_ready = analysis.translate(to='en')

        if analysis_ready.sentiment.polarity > 0: 
            return 'positivo'
        elif analysis_ready.sentiment.polarity == 0: 
            return 'neutral'
        else: 
            return 'negativo'
    except:
        return None
```

La traducci√≥n de nuestros tweets, previamente descrita, se ejecuta conectando con la API de Google. Esta API tiene una limitaci√≥n de peticiones diarias, por lo que dado el volumen de datos de nuestro dataframe, necesitamos una versi√≥n abreviada para que sea posible ejecutar el c√≥digo de la traducci√≥n y proseguir con el an√°lisis de sentimientos de nuestros tweets. Hacemos una copia reducida de nuestro dataframe y creamos una nueva columna que registre el "sentimiento" de los tweets utilizando la funci√≥n que creamos antes. Visualizamos el resultado del nuevo dataframe para comprobar que se ha creado correctamente.

```python
short_df = df[:500].copy()
short_df['Sentimiento'] = short_df['Texto'].apply(analize_tweet_sentiment)
short_df
```

Por √∫ltimo, hacemos una agregaci√≥n de la columna "Sentimiento" para ver cu√°ntos tweets hay en cada una de nuestras tres categor√≠as y sacar cu√°l es el sentimiento predominante de los tweets recolectados, as√≠ como el porcentaje que representa este resultado frente al total de tweets. Para ver los resultados agrupados por categor√≠as, realizamos un diagrama de barras.

```python
df_grouped_sent = short_df.groupby(['Sentimiento'])
count_sent = df_grouped_sent.Sentimiento.count()
print("El sentimiento predominante de los tweets es: {}".format(count_sent.idxmax()))                                                                     
print("El porcentaje de tweets con este sentimiento fue del: {}%".format(count_sent.max()*100/count_sent.sum()))
```
